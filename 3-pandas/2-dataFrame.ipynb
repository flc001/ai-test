# DateFrame

### DataFrame就是Series组成的字典 共用同一个索引 从一位扩展到多维

```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import matplotlib.pyplot as plt

```

```python
'''
    DateFrame的创建
    最常用的方法是传递一个字典来创建 DataFrame以字典的键作为每一列的名称
    以字典的值作为每一列
    
    shift+tab 查看函数的参数
'''
#有个copy=False 是否复制 如果是True  不使用引用 而是在内存中开辟一个空间 一个副本
df = DataFrame({
    'height':[181,166,172,188],
    'age':np.random.randint(18,24,size = 4),
    'sex':['男','女','女','男'],
    'weight':[90,50,48,75] #没有可以 但是有 就必须是4个
},index=['瓦里安','吉安娜','奥蕾莉亚','卡德加'],
columns = ['height','age','sex','weight']
)
#weight没有数据
# df.index = [1,2,3,4]#可以使用df.index赋值

#检索列 返回值
age = df['age']

display(df,age,type(age))
#类型 pandas.core.series.Series

#对于DataFrame 列名 就相当于属性
#DataFrame是统计数据 用的表名 某一个事物的属性 每个属性对于dataFrame的列名


```

```python
#检索行 行不能直接检索   
'''
    .ix[] 来进行行索引
    .loc[] 加index来进行索引
    .iloc[] 加整数 来进行索引
'''
#检索多个时  需要一个中括号 作为条件 
#如果检索多行 返回的数据 是dataFrame 不能切片
display(df.loc[['吉安娜','奥蕾莉亚']])

#切片效果 左闭右闭
display(df.loc['瓦里安':'奥蕾莉亚'])


display(df.iloc[1:3])

```

```python
#dataFrame自身有bug 索引是汉字 有时无法检索结果 个别汉字

df['age']['吉安娜'] = '40'
#这个是提醒 检验我们copy 没报错
display(df)


df.loc['瓦里安']['height'] = 195
#检索行的时候 参数可以多个列无法完成这样的操作
df.loc['瓦里安','weight'] = 105
# df.values[0,2] 行,列 这种方式也可以
display(df)

#使用中括号时 
#索引表示列索引 切片表示行切片 df['height':'age'] 结果一个都没  height和age是列索引
```

## dataFrame的运算 
> 在运算中 自动对齐 不同索引的数据

> 如果索引不对称 则补NaN

```python
#4行4列
df1 = DataFrame(
    np.random.randint(50,100,size=(4,4)),
    index=['张三','李四','王五','马六'],
    columns = ['语文','数学','英语','物理']
)
display(df1)
#月考二 转入新同学 变成五个人 size里面是行和列
df2 = DataFrame(
    np.random.randint(50,100,size=(5,4)),
    index=['张三','李四','王五','马六','老刘'],
    columns = ['语文','数学','英语','物理']
)
display(df2)
#求月考1 2的和
```

```python
display(df1+df2)#老刘的分数为NaN 
#避免这种情况 使用pandas methods
'''
    python的加减乘除不精细 可以避免 NaN
'''
df3 = df1.add(df2,fill_value=0)
display(df3)#索引不对应 不管是行还是列 
```

```python
#series和dataframe
s1 = df3['数学']
#所有成绩变成NAN 考试名字也出现在学科上面  张三	数学	李四	物理
# df3+s1
#一维和二维相加 把一维的索引  也就是人名 变成各列的列名


s2 = df3.loc['李四']
# s2 相加 就看 列索引是否相同
#广播模式
df3+s2
#add sub mul div floordiv整除 mod求余 pow
```

```python
#axis=0 以列为单位操作 参数必须是列 对所有的列都有效
#axis=1 以行为单位操作 参数必须是行 对所有的行都有效
#也能给index或者columns 默认是columns

#如果ss是列 就得以行为单位 提取一行
ssi = df3.loc['张三']

#如果ss提取一列
ssc = df3['数学']#张三	数学	李四	物理	王五	老刘	英语	语文	马六 全是NaN

# df3.add(ss)
```

```python
df3.add(ssi,axis = 0)#等于0 看行索引 列不会出现错位 
# df3.add(ssc,axis = 0)

#等于0 以列为单位 操作
```
