# Pandas的数据结构

```python
import matplotlib.pyplot as plt

#使用魔法指令 嵌入进去
%matplotlib inline 
```

```python
img1 = plt.imread('img1.png')
plt.imshow(img1)
```

```python
#竖直方向进行颠倒 
img2 = img1[::-1]
plt.imshow(img2)
```

```python
#竖直和水平方向进行颠倒 
img3 = img1[::-1,::-1]
plt.imshow(img3)
```

```python
#颜色颠倒 红绿蓝 之前越红 现在越蓝
img4 = img1[:: , :: , ::-1]
plt.imshow(img4)
```

```python
# img5 = img1[::5,::5]#代表 每隔五个 取一个数据 这不叫马赛克 数据都少了
# plt.imshow(img5)

#发现出现很多点 
img1[::5,::5] = 10
plt.imshow(img1)
```
